```python
import numpy as np
import matplotlib.pyplot as plt
import torch
from torch import nn
from torch.autograd import Variable
from torch.nn.parameter import Parameter
from importlib import reload
def chol(cov):
    """
    cov = a@a.T
    return a
    """
    return torch.linalg.cholesky(cov)



from VAE_project_lib.sym_blk_tridiag_inv import *
from VAE_project_lib.blk_tridiag_chol_tools import *
def bidiag_builder(AA,BB,lower=False):
    """
    Build a bi-diagonal block matrix given the blocks
    AA is the diagonal blocks, [T,n,n]
    BB is the off-diagonal blocks, [T-1,n,n]
    """
    import torch
    T, n, _ = AA.shape  # 获取T和n的值
    N = T * n  # 最终矩阵的大小    
    # 初始化一个大小为 N x N 的零矩阵
    matrix = torch.zeros(N, N, dtype=AA.dtype, device=AA.device)    
    # 填充对角块
    for i in range(T):
        start_idx = i * n
        end_idx = start_idx + n
        matrix[start_idx:end_idx, start_idx:end_idx] = AA[i]    
    # 填充次对角块
    for i in range(T - 1):
        start_idx = i * n
        if lower:
            # 在下对角线上填充
            matrix[start_idx+n:start_idx+2*n, start_idx:start_idx+n] = BB[i]
        else:
            # 在上对角线上填充
            matrix[start_idx:start_idx+n, start_idx+n:start_idx+2*n] = BB[i]    
    return matrix
def tridiag_builder(AA,BB,lower=True):
    """
    Build a tri-diagonal symmetric block matrix given the blocks
    AA is the diagonal blocks, [T,n,n]
    BB is the off-diagonal blocks, [T-1,n,n]
    """
    import torch
    T, n, _ = AA.shape  # 获取T和n的值
    N = T * n  # 最终矩阵的大小    
    # 初始化一个大小为 N x N 的零矩阵
    matrix = torch.zeros(N, N, dtype=AA.dtype, device=AA.device)    
    # 填充对角块
    for i in range(T):
        start_idx = i * n
        end_idx = start_idx + n
        matrix[start_idx:end_idx, start_idx:end_idx] = AA[i]    
    # 填充次对角块
    for i in range(T - 1):
        start_idx = i * n

        if lower:
        # 在下对角线上填充B
            matrix[start_idx+n:start_idx+2*n, start_idx:start_idx+n] = BB[i]
        # 在上对角线上填充B.T
            matrix[start_idx:start_idx+n, start_idx+n:start_idx+2*n] = BB[i].T
        else:
            matrix[start_idx+n:start_idx+2*n, start_idx:start_idx+n] = BB[i].T
            matrix[start_idx:start_idx+n, start_idx+n:start_idx+2*n] = BB[i]
    return matrix
```
